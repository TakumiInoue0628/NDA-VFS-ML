```python
import os
import numpy as np

from config import MRI8_NVF4_10mm_0p0mmVVVV as cfg

os.chdir('../')
from lib.methods import *
from lib.figure import NLP

os.chdir('../')
```

## Load data

```python
_, video_data_list, video_t_data_list = LoadCSVandVIDEOS(
                                                        csv_path=cfg.CSV_FILE_DIR+cfg.DATA_NAME+'.csv', 
                                                        data_name_list=['t', 'mic'], 
                                                        sample_span=(160000, 170000), 
                                                        videos_path_list=[
                                                                        cfg.VIDEO_FILE_DIR+'VF/'+cfg.DATA_NAME+'.avi',
                                                                        cfg.VIDEO_FILE_DIR+'FVF/'+cfg.DATA_NAME+'.avi',
                                                                        ]
                                                        )
video_vf_data = video_data_list[0]
video_fvf_data = video_data_list[1]
video_vf_t_data = video_t_data_list[0]
video_fvf_t_data = video_t_data_list[1]
```

## Figure

```python
fig = NLP({
            'font.family':'Times New Roman',
            'mathtext.fontset':'stix',
            'font.size':30,
            'xtick.direction':'in',
            'ytick.direction':'in',
            'axes.linewidth':1.5,
            'xtick.major.size':8,
            'ytick.major.size':8,
            })

fig.figure03(data=video_vf_data, t_data=video_vf_t_data, scanline=True, position=[50, 40], width=40, c_line='r', lw_line=4, start=80, step=30, start_t_is_0=True,
                    title='Vocal fold',
                    figsize=(20, 2), n_shots=8, wspace=0.1, aspect=1.3,
                    gamma=0.8,
                    panel = '(a)', panel_xy=(0.12, 1.), panel_fontsize=30,
                    save_png=None, save_eps=None)

fig.figure03(data=video_fvf_data, t_data=video_fvf_t_data, scanline=True, position=[70, 80], width=70, c_line='r', lw_line=4, start=90, step=30, start_t_is_0=True,
                    title='Ventricular fold',
                    figsize=(20, 2), n_shots=8, wspace=0.1, aspect=1.0,
                    gamma=0.5,
                    panel = '(b)', panel_xy=(0.12, 1.), panel_fontsize=30,
                    save_png=None, save_eps=None)
```
